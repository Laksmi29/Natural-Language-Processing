```python
import pandas as pd
import numpy as np
import re
import re as reg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Baca data hasil scrapping dari excel
data = pd.read_excel('JavaHeritageHotel.xlsx')
data
```

```python
#Preprocessing
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

slangs={'yg':'yang', 'tdk':'tidak', 'pd':'pada', 'mlh':'malah', 'jgn':'jangan', 'jg':'juga', 'tp':'tapi', 'blkg': 'belakang', 'dr':'dari', 'klo':'kalo', 'lg':'lagi', 'dlm':'dalam', 'dgn':'dengan', 'g':'tiddak', 'aj':'saja',
       'blh':'boleh', 'jg':'juga', 'jd':'jadi', 'lgsg':'langsung'}
processed_comments = []

for sentence in data['customer_isi_komen']:
    # Remove all the special characters
    processed_comment = re.sub(r'\W', ' ', str(sentence))
    
    # Converting to Lowercase
    processed_comment = processed_comment.lower()
    
    #Remove number 
    processed_comment = re.sub(r'\d+', ' ', processed_comment)

    # remove all single characters
    processed_comment = re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_comment)
    
    #remove duplicate character
    pattern=reg.compile(r"(.)\1{1,}",reg.DOTALL)
    processed_comment=pattern.sub(r"\1",processed_comment)


    #Corrected Slang words
    words = processed_comment.split()
    rfrm=[slangs[word] if word in slangs else word for word in words]
    processed_comment= " ".join(rfrm)

    #remove stopword
    factory = StopWordRemoverFactory()
    more_stopword = ['tak', 'jd', 'per', 'nya', 'terjemah', 'diterjemahkan', 'oleh', 'gogle', 'google' ,'nan'] #menambahkan stopword
    stopwords = factory.get_stop_words() + more_stopword
    temp = [t for t in re.findall(r'\b[a-z]+-?[a-z]+\b',processed_comment) if t not in stopwords]
    processed_comment = ' '.join(temp)
    
    #stemming
    stemmer = StemmerFactory().create_stemmer()
    processed_comment = stemmer.stem(processed_comment)
    
    # Substituting multiple spaces with single space
    processed_comment = re.sub(r'\s+', ' ', processed_comment, flags=re.I)

    processed_comments.append(processed_comment)
    
#output data Preprocessing
processed_comments
```

```python
import nltk
```

```python
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize

k = processed_comment
kalimat = nltk.sent_tokenize(k)
print ("Kalimat Awal: ", kalimat)
hasil = nltk.word_tokenize(k)
print("Hasil Tokenizing: ", hasil)
```

# Vektorisasi TF

```python
from sklearn.feature_extraction.text import CountVectorizer
tf_vec = CountVectorizer(max_df=1.0, min_df=1)
tf = tf_vec.fit_transform(hasil)

#Hasil representasi
tf_term = tf_vec.get_feature_names()
print(tf_vec.get_feature_names())
matrix = tf.toarray()
print(matrix)
```

# Modelling dengan 5 Topik

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA

topik = 5 #utk mendptkn jml topik terbaik perlu trial
lda = LDA(n_components = topik, learning_method = 'batch', random_state = 0).fit(tf)
lda
```

```python
#training LDA
vsm_topik = lda.transform(tf)
#tampilkan hasil
print(vsm_topik.shape)
vsm_topik
```

```python
#Tampilan nilai-nilai setiap fitur
print(lda.components_)
```

```python
#Hasil Label Topic Model setiap Dokumen
import numpy as np
topics = np.argmax(vsm_topik, axis = 1)
topics
```

```python
#Cetak Word Fitur dgn Nilai Tertinggi pada Setiap Topik
n_top = 4 #Jml Fitur Tertinggi
topik_word = {}

for topic, comp in enumerate(lda.components_):
    word_idx = np.argsort(comp)[::-1][:n_top]
    
    # store the words most relevant to the topic
    topik_word[topic] = [tf_vec.get_feature_names()[i]+' '+str(comp[i]) for i in word_idx]
```

```python
for topic, words in topik_word.items():
    print('Topic: %d' % topic)
    print(' %s' % ', '.join(words))
```
